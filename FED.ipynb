## Influência das Condições Econômicas nas Decisões de Taxas de Juros do Federal Reserve

### Objetivo
Analisar como o crescimento econômico, a taxa de desemprego e a taxa de inflação afetam as decisões de taxas de juros do Federal Reserve e prever futuras decisões de taxas de juros com base nas tendências históricas e nos dados econômicos atuais.

### Dados Disponíveis
1. Taxa de fundos federais (1954 - presente)
2. Produto Interno Bruto Real (ajustado sazonalmente)
3. Taxa de desemprego (ajustada sazonalmente)
4. Taxa de inflação (variação mensal no Índice de Preços ao Consumidor excluindo alimentos e energia)

### Fontes de Dados
- Taxa de juros: Federal Reserve Bank of St. Louis
- PIB: US Bureau of Economic Analysis
- Taxa de desemprego e preços ao consumidor: US Bureau of Labor Statistics

### Metodologia
1. **Coleta de Dados**

2. **Pré-processamento de Dados**
   - Limpeza dos dados (tratar valores ausentes, remover duplicatas, etc.).

3. **Análise Exploratória de Dados (EDA)**
   - Estatísticas descritivas (média, mediana, desvio padrão, etc.).
   - Visualização de dados (gráficos de linha, histogramas, scatter plots).

4. **Análise Temporal**
   - Estudo de séries temporais para identificar tendências e padrões sazonais.
   - Análise de correlação entre as variáveis econômicas e a taxa de fundos federais.

5. **Modelagem Preditiva**
   - Construção de modelos de regressão para prever a taxa de fundos federais com base nas variáveis econômicas.
   - Validação dos modelos utilizando métricas de desempenho (RMSE, MAE, etc.).

6. **Previsão de Taxas de Juros**
   - Utilização dos modelos validados para prever futuras decisões de taxas de juros do Federal Reserve.
   - Análise de cenários para diferentes condições econômicas.

7. **Interpretação dos Resultados**
   - Identificação dos principais fatores econômicos que influenciam as decisões de taxas de juros.
   - Discussão sobre como a política de juros mudou ao longo do tempo.

```python
# Importando as bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.impute import SimpleImputer

sns.set(style="darkgrid")

dados_combinados = pd.read_csv('/content/drive/MyDrive/Portfolio/FED/FED.csv')

dados_combinados['date'] = pd.to_datetime(dados_combinados[['Year', 'Month', 'Day']])

X = dados_combinados[['Real GDP (Percent Change)', 'Unemployment Rate', 'Inflation Rate']]
y = dados_combinados['Effective Federal Funds Rate']

# Imputação de valores ausentes usando a média
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)
y = imputer.fit_transform(y.values.reshape(-1, 1))

# Dividindo os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Construção do modelo de regressão linear
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# Previsão no conjunto de teste
y_pred = modelo.predict(X_test)

# Validação do modelo utilizando métricas de desempenho
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'RMSE: {rmse}')
print(f'MAE: {mae}')
print(f'R²: {r2}')

# Visualização das previsões
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], '--', color='red')
plt.xlabel('Valores Reais')
plt.ylabel('Valores Previstos')
plt.title('Valores Reais vs. Valores Previstos')
plt.show()

coeficientes = pd.DataFrame(modelo.coef_.reshape(-1, 1),
                           index=['Real GDP (Percent Change)', 'Unemployment Rate', 'Inflation Rate'],
                           columns=['Coeficiente'])
# modelo.coef_ has shape (1, 3) and needs to be reshaped to (3, 1)
# to match the expected shape of the DataFrame with 3 rows (index) and 1 column.
```

# Previsão de Taxas de Juros
Utilização dos Modelos Validados
Vamos usar o modelo de regressão linear que construímos para prever futuras decisões de taxas de juros do Federal Reserve.

Previsão com os Modelos Validados: Vamos criar cenários hipotéticos para diferentes variáveis econômicas e prever as taxas de juros correspondentes.

```python
# Cenário 1: Crescimento Econômico Alto, Desemprego Baixo, Inflação Baixa
cenario_1 = np.array([[3.0, 4.0, 1.5]])  # [PIB, Desemprego, Inflação]
predicao_cenario_1 = modelo.predict(cenario_1)
print(f'Previsão de Taxa de Juros (Cenário 1): {predicao_cenario_1[0][0]:.2f}%')

# Cenário 2: Crescimento Econômico Baixo, Desemprego Alto, Inflação Alta
cenario_2 = np.array([[0.5, 9.0, 3.5]])  # [PIB, Desemprego, Inflação]
predicao_cenario_2 = modelo.predict(cenario_2)
print(f'Previsão de Taxa de Juros (Cenário 2): {predicao_cenario_2[0][0]:.2f}%')

# Cenário 3: Crescimento Econômico Moderado, Desemprego Moderado, Inflação Moderada
cenario_3 = np.array([[2.0, 5.5, 2.0]])  # [PIB, Desemprego, Inflação]
predicao_cenario_3 = modelo.predict(cenario_3)
print(f'Previsão de Taxa de Juros (Cenário 3): {predicao_cenario_3[0][0]:.2f}%')
```

# Análise de Cenários
Vamos criar diferentes cenários econômicos e prever as taxas de juros para cada um, visualizando os resultados.

```python
# Configurando os cenários econômicos
cenarios = {
    "Crescimento Econômico Alto, Desemprego Baixo, Inflação Baixa": [3.0, 4.0, 1.5],
    "Crescimento Econômico Baixo, Desemprego Alto, Inflação Alta": [0.5, 9.0, 3.5],
    "Crescimento Econômico Moderado, Desemprego Moderado, Inflação Moderada": [2.0, 5.5, 2.0],
}

# Previsão e armazenamento dos resultados
resultados_previsao = {}
for descricao, valores in cenarios.items():
    predicao = modelo.predict(np.array([valores]))
    resultados_previsao[descricao] = predicao[0][0]

# Exibindo os resultados
for descricao, taxa in resultados_previsao.items():
    print(f"{descricao}: Previsão de Taxa de Juros = {taxa:.2f}%")
```

# Visualização dos Cenários
Para visualizar melhor as previsões de taxa de juros para cada cenário, vamos criar um gráfico.

```python
descricao_cenarios = list(resultados_previsao.keys())
taxas_previsao = list(resultados_previsao.values())

# Criando o gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(descricao_cenarios, taxas_previsao, color=['blue', 'orange', 'green'])
plt.xlabel('Cenários Econômicos')
plt.ylabel('Previsão de Taxa de Juros (%)')
plt.title('Previsão de Taxas de Juros para Diferentes Cenários Econômicos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
